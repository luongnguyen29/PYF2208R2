## Review: Vẽ kim tự tháp


```python
#Nhập vào số lượng tam giác cần vẽ. Sử dụng turtle để vẽ ra kim tự tháp như hình bên trên
# Mỗi tam giác có cạnh > tam giác trước 10

import turtle
pen = turtle.Turtle()

total = int(input("Nhập số lượng tam giác cần vẽ: "))
count = 1
i = 10
while count <= total:
    pen.circle(10*count, steps=3)
    count+=1
    
turtle.done()
```

## for kết hợp với hàm range()

### Trường hợp 1 tham số

```python
# In ra các số tự nhiên từ 1 đến 10
```

```python
#While loop
count = 1
while count < 11:
    print(count)
    count+=1
```

```python
for i in range(11):
    print(i)
```

### Ví dụ: In các số lẻ từ 0 đến 100

### Trường hợp 2 tham số

```python
for i in range(5,11): #auto cộng 1 hở
    print(i)
```

### Ví dụ: In các số chia hết cho 7 từ a đến b

```python
for i in range(1000,5000):
    if i%7==0:
        print(i)
```

### Trường hợp 3 tham số

```python
for i in range(10, -1, -1): #Nếu ko có số 3 thì bước nhảy auto = 1, range(Start-optional(default to 0), end-required, step-optional default 1)
    print(i)
```

### Ví dụ: Biểu diễn cấp số cộng

```python
#Nhập vào số hạng đầu và công sai của cấp số cộng, sau đó in ra 10 số hạng đầu tiên của cấp số cộng đó
a = int(input("Số thứ nhất: "))
b = int(input("công sai: "))
c = int(input("Nhập số lần muốn lặp: "))
for i in range(a,a+c*b+1,b):
        print(i)
```

## lặp qua một chuỗi

```python
string = "Hello Python Fundamental"
for i in range(len(string)): #Hàm len cho biết string có bao nhiêu ký tự
    print(f'Index: {i}, value: {string[i]}') 
```

```python
#Cách 2: duyệt string và chỉ cho ra ký tự, ko cần dùng index - c là stand for character.
string = "Hello Python Fundamental"
for c in string: #c = character
    print(c)
```

### Ví dụ: Đếm số lượng nguyên âm trong chuỗi

```python
string = "hello python fundamental"
count = 0
for c in string:
    if c=='a' or c=='u' or c=='e' or c=='i' or c=='o':
        count+=1
print("Số nguyên âm trong chuỗi: ",count, end='')
```

```python
string = "hello python fundamental"
n = 0
for c in string:
    if c in {"u","o","e","i","a"}:
        n+=1
print(n)
```

## Lặp qua một danh sách chuỗi

```python
name = ["Luong", "Thien", "Nhan"]
for i in range(len(name)):
    print(name[i])
```

## Vòng lặp lồng nhau

```python
# In bảng cửu chương
for i in range(1,10):
    for j in range(1,11):
        print(i*j, end=' ')
    print() # cái chỗ này để xuống dòng thôi
```

```python
# In list số từ 1 - 100 
for i in range(0,10):
    for j in range(1,100,10):
        print(j+i, end=" ")
    print()
```
