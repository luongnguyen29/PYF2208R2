## Viết chương trình nhập vào điểm TB 3b A, B, C. Cho biết bạn nào có điểm lớn nhất.


```python
a = float(input("A's average score: "))
b = float(input("B's average score: "))
c = float(input("C's average score: "))
if a>b and a>c:
    print(f'The highest average score is A {a}')
elif b>c:
    print(f'The highest average score is B {b}')
else: 
    print(f'The highest average score is C {c}')
```

## Cấu trúc lặp while

### Ví dụ: In ra tất cả số chẵn từ 0 đến n

```python
num = int(input("Please input n: "))
count = 0 #Gán giá trị 0 cho count
while count <= num: #while miễn là
    print(count)
    count += 2
```

### Ví dụ: In ra tất cả số lẻ từ n về 1

```python
num1 = int(input("Please input n1: "))
count = 0 #Gán giá trị 0 cho count
while 1<= count <= num1: #while miễn là
    print(count)
    count -= 2
```

## Input: abcdfege, Output: egefdcba

```python
string = str(input("Please input string: "))
index = len(string)-1
while index >= 0:
    print(string[index],end=(''))
    index -=1
```
